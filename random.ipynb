# RANDOM NUMBERS IN NUMPY

```python
from numpy import random
import numpy as np
```

```python
x = random.randint(100)
print(x)
```

```python
# random float from 0 to 1

x = random.rand()
print(f"{x:.2f}")
```

## Generate random array

```python
arr = random.randint(100, size=5)
print(arr)
```

```python
arr = random.randint(100,size=(3,5))
print(arr)
```

```python
arr = random.rand(5)
print(arr)
```

```python
arr = random.rand(3,5)
print(arr)
```

## Generate random values from array

```python
x = random.choice([3, 6, 8, 1, 9])
print(x)
```

```python
arr = random.choice([4, 7, 12, 9, 2],size=(3, 5))
print(arr)
```

# RANDOM DATA DISTRIBUTION

```python
arr = random.choice([2,3,4,5],size=(10),p=[0.5,0.2,0.3,0])
print(arr)
```

```python
arr = random.choice([4,5,6,7],size=(3,5),p=[0.3,0.4,0.3,0])
print(arr)
```

# RANDOM PERMUTATIONS

```python
arr = np.array([1, 2, 3, 4, 5, 6])
newarr = random.shuffle(arr)
print(arr)
print(newarr)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
newarr = random.permutation(arr)
print(arr)
print(newarr)
```

# NORMAL(Gaussian) Distribution

```python
x = random.normal(size=(2,3))
print(x)
```

```python
x = random.normal(loc=1,scale=2,size=(2,3))
print(f"{x}")
```

```python
sns.displot(random.normal(size=(1000)),kind="kde")
```

# BINOMIAL DISTRIBUTION

```python
arr = random.binomial(n=10,p=0.5,size=10)
print(arr)
```

```python
arr = random.binomial(n=10,p=0.5,size=1000)
# print(arr)
sns.displot(arr)
```

## Diference between Normal and Binomial distribution

```python
data = {
    'normal':random.normal(loc=50,scale=5,size=1000),
    'binomial': random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data,kind="kde")
```

# POISSON DISTRIBUTION

```python
x = random.poisson(lam=2,size=10)
print(x)
```

```python
x = random.poisson(lam=2,size=1000)
print(x)
sns.displot(x)
```

## Difference between normal and poisson distribution

```python
data = {
   "normal" :  random.normal(loc=50,scale=7,size=1000),
   "poisson" : random.poisson(lam=50,size=1000)
}

sns.displot(data,kind="kde")
```

## Difference between binomial and poisson distribution

```python
data = {
    "binomial": random.binomial(n=1000, p=0.01, size=1000),
    "poisson": random.poisson(lam=10,size=1000)
}

sns.displot(data=data,kind="kde")
```

# UNIFORM DISTRIBUTION

```python
x=random.uniform(low=1,high=10,size=10)
print(x)
```

```python
x = random.uniform(size=(2,4))
print(x)
```

```python
x = random.uniform(size=100)
sns.displot(x,kind="kde")
```

# LOGISTIC DISTRIBUTION

```python
x = random.logistic(loc=1,scale=2,size=(2,3))
print(x)
```

```python
x = random.logistic(size=1000)
sns.displot(data=x,kind="kde")
```

## Difference Between Logistic and Normal Distribution

```python
data = {
    "normal": random.normal(scale=2,size=1000),
    "logistic": random.logistic(size=1000),
}

sns.displot(data, kind="kde")
```
